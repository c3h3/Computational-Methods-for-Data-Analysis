# Video: W03_L07_P01

## [Marine Mammals Spectrogram](http://www.pmel.noaa.gov/acoustics/specs_whales.html)

# Initialize Setting: 

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
L = 10
n = 2048
t2 = np.linspace(0,L,n+1)
t = t2[:-1]
k = (2*np.pi/L)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.ifftshift(k)
````

# Create and Plot the Signal: 

````python
#Sample_Signal
S = (3*np.sin(2*t) + 0.5*np.tanh(0,5*(t-3)) + 0.28*np.exp(-(t-4)**2) \
+ 1.5*np.sin(5*t) + 4*np.cos(3*(t-6)**2))/10 + (t/20)**3
````

````python
plt.plot(t,S)
````

# Compute and Plot FFT of the Signal: 

````python
St = np.fft.fft(S)
````

````python
plt.subplot(2,1,1)
plt.plot(t,S)
plt.subplot(2,1,2)
plt.plot(ks,np.abs(np.fft.fftshift(St)))
````

# Zoom In the second figure: 

````python
plt.subplot(2,1,1)
plt.plot(t,S)
plt.subplot(2,1,2)
plt.plot(ks,np.abs(np.fft.fftshift(St)))
plt.xlim(-50, 50)
````

# Re-Scaling and ZoomIn second figure: 

````python
plt.subplot(2,1,1)
plt.plot(t,S)
plt.subplot(2,1,2)
plt.plot(ks,np.abs(np.fft.fftshift(St))/max(np.abs(np.fft.fftshift(St))))
plt.xlim(-50, 50)
````

# Create Windows with Different Sizes:

````python
width_list = [10.0,1.0,0.2]

for i in range(3):
    f = np.exp(-width_list[i]*(t-4)**2)
    plt.subplot(3,1,i+1)
    plt.plot(t,S,"k",t,f,"m")
    plt.ylim(-1.25, 1.25)
````
